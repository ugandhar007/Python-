```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd

app = dash.Dash()


# https://dash.plot.ly/dash-core-components/dropdown
# We need to construct a dictionary of dropdown values for the years
dates=[]
for i in create_df['Job_comp_Time'].unique():
    d=str(i)
    dates.append(d[:10]) 
app.layout = html.Div([
    dcc.Graph(id='graph'),
    dcc.Dropdown(id='t1',options=[{'label':i,'value':i}for i in dates],value='2019-11-01'),
    dcc.Dropdown(id='t2',options=[{'label':i,'value':i}for i in dates],value='2019-11-05')
])
@app.callback(Output('graph','figure'),
              [Input('t1','value'),
              Input('t2','value')])
def update_figure(t1,t2):
    mask=(create_df['Job_comp_Time'] > '2019-11-01' & create_df['Job_comp_Time'] <= '2019-11-01')
    f_df=create_df.loc(mask)
    p4=np.count_nonzero(f_df['Product']=='Product 4', axis=0)
    p3=np.count_nonzero(f_df['Product']=='Product 3', axis=0)
    p2=np.count_nonzero(f_df['Product']=='Product 2', axis=0)
    p1=np.count_nonzero(f_df['Product']=='Product 1', axis=0)
    p5=np.count_nonzero(f_df['Product']=='Product 5', axis=0)
    traces = []
    traces.append(go.Pie(labels=["Product 1","Product 2","Product 3","Product 4","Product 5"],values=[p1,p2,p3,p4,p5]))

        #dcc graph takes input in the form of dictionary
    return {'data':traces }
if __name__=='__main__':
   app.run_server()
   


```
